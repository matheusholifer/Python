```python
'''
    4) Tratamento de Nulos: Identifique valores ausentes no dataset. Escolha entre removê-los ou preenchê-los com a média da coluna usando o método .fillna().
'''
```

```python
import seaborn as sns

df = sns.load_dataset('tips')

print(df.head())
```

```python
import numpy as np

# Consultando o total de nulos por coluna usando total_nulos = df['size'].isnull().sum()
# Foi constatado que não existem valores nulos no dataset 'tips'
# Então irei criar uma cópia do dataset com valores faltantes
novo_dataframe = df[(df['total_bill'] > 15) & (df['total_bill'] < 25)  & (df['tip'] > 1.5)].copy()

# Sempre que a gorjeta for menor que $1.5, o dado se transforma em NaN
df.loc[df['tip'] < 1.5, 'tip'] = np.nan

print(novo_dataframe)
```

```python
# Soma os valores NaN na coluna 'tip'
total_nulos = df['tip'].isnull().sum()

print(total_nulos) # Saída: 17
```

```python
# Cálculo da média da coluna 'tip'
media_gorjetas = df['tip'].mean()

# Preenche os nulos da coluna com a média calculada
df['tip'] = df['tip'].fillna(media_gorjetas)
```

```python
# Soma os valores NaN na coluna 'tip'
total_nulos = df['tip'].isnull().sum()

print(total_nulos) # Saída: 0
```

```python
'''
    10) Ordenação: Escolha uma coluna numérica e ordene o DataFrame do maior para o menor valor (.sort_values()). Salve apenas os 10 maiores resultados em uma nova variável.
'''
```

```python
# Utilizamos seaborn para seleção de datasets internos
import seaborn as sns

# Carrega o dataset interno 'penguins' 🐧
df = sns.load_dataset('penguins')

df.head() # Mostra as 5 primeiras linhas do dataset
```

```python
'''
Coluna              Descrição
species             A espécie do pinguim (Adélie, Chinstrap ou Gentoo).
island              A ilha onde foi coletado (Biscoe, Dream ou Torgersen).
bill_length_mm      O comprimento do bico (em milímetros).
bill_depth_mm       Aprofundidade/altura do bico (em milímetros).
flipper_length_mm   O comprimento da nadadeira (em milímetros).
body_mass_g         A massa corporal (em gramas).
sex                 O sexo do animal (Male ou Female).
'''
```

```python
# Ordenar da maior para a menor massa corporal
# ascending=False garante a ordem decrescente
# .head(10) seleciona apenas os 10 primeiros
top_10_pesados = df.sort_values(by="body_mass_g", ascending=False).head(10)

# Exibir o resultado
print(top_10_pesados)
```

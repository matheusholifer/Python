```python
'''
    06) Primeiros Gráficos: Use o Matplotlib para criar um gráfico de barras com os dados agrupados no exercício anterior. Em seguida, use o Seaborn (sns.histplot) para visualizar a distribuição de uma variável numérica.
'''
```

```python
# Importa a biblioteca seaborn e configura o dataset 'titanic'
import seaborn as sns

df = sns.load_dataset('titanic')

# Mostra as 5 primeiras linhas do dataset
df.head()
```

```python
# Importação de pandas e utilização do método groupby
import pandas as pd

# Agrupamento por classe, calculando a arrecadação da tarifa
soma_ticket = df.groupby('class')['fare'].sum()

print(soma_ticket)
```

```python
# Criação do Gráfico de Barras (Bar Chart) usando matplotlib
# importa as bibliotecas matplotlib e numpy para a criação do gráfico
import matplotlib.pyplot as plt
import numpy as pd

# Definindo o tamanho da figura
plt.figure(figsize=(8, 5))

# Sintaxe para a ccriação do gráfico: plt.bar(eixo_x, eixo_y)
plt.bar(soma_ticket.index, soma_ticket.values, color='skyblue')

# Adicionando títulos e rótulos
plt.title('Faturamento por Classe')
plt.xlabel('Class')
plt.ylabel('Fare')

plt.show()
```

```python
# Configuração visual do Seaborn (Opicional)
sns.set_theme(style="whitegrid")

# Criando o Histograma
plt.figure(figsize=(10, 6))

# kde=True adiciona a linha de densidade suave sobre as barras
sns.histplot(data=df, x='fare', kde=True, color='teal', bins=30)

# Personalização
plt.title('Distribuição das Tarifas (Fare)')
plt.xlabel('Preço da Tarifa')
plt.ylabel('Frequência')

plt.show()
```

```python
# Importação e configuração do dataset a ser utilizado
import seaborn as sns

df = sns.load_dataset('tips')
```

```python
# Visualizando a estrutura do dataset 'tips' - gorjetas
import pandas as pd

print(df.head())
```

```python
# Criando um novo dataframe
# Apenas as pessoal que pagaram entre $15.0 e $25.0 na refeiçaõ, e deixaram uma gorjeta maior que $1.5
novo_dataframe = df[(df['total_bill'] > 15) & (df['total_bill'] < 25)  & (df['tip'] > 1.5)].copy()

# Visualizando o novo dataframe
print(novo_dataframe)
```

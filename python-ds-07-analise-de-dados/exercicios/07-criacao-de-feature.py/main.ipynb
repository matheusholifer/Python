```python
# Importação da biblioteca seaborn
import seaborn as sns

# Carrega o dataset interno 'titanic'
df = sns.load_dataset('titanic')

df.head() # Mostra as 5 primeiras linhas do 'titanic'
```

```python
# Feature: FamilySize
import pandas as pd

# Cria a feature usando Operações Aritméticas Simples
df['FamilySize'] = df['sibsp'] + df['parch'] + 1

df.head()
```

```python
# Featrure: IsAlone
import numpy as np

# Se FamilySize for 1, IsAlone recebe 1, caso contrário recebe 0
df['IsAlone'] = np.where(df['FamilySize'] > 1, 0, 1)

df.head()
```

```python
# Feature: AgeGroup
# Usando o método .apply() com Funções
def categorizar_idade(idade):
    if idade < 12:
        return 'Criança'
    elif idade < 18:
        return 'Adolescente'
    else:
        return 'Adulto'

df['AgeGroup'] = df['age'].apply(categorizar_idade)

df.head()
```

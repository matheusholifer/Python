```python
'''
    8) Correlação Visual: Utilize o Seaborn para criar um heatmap (sns.heatmap) que mostre a correlação entre todas as variáveis numéricas do seu dataset.
'''
```

```python
# Importa o seaborn
import seaborn as sns

# Configura o dataset interno 'titanic' para utilização
df = sns.load_dataset('titanic')

# Mostra as 5 primeiras linhas do dataset
df.head()
```

```python
import matplotlib.pyplot as plt

# Selecionando apenas as variáveis numéricas
corr = df.corr(numeric_only=True)

# Configurando o tamanho da figura
plt.figure(figsize=(12, 8))

# Criando o heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Exibindo o gráfico
plt.title('Matriz de Correlação das Variáveis')
plt.show()
```

```python
# Lidando com a redundânccia
import numpy as np

# Criando uma máscara para esconder o triângulo superior
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt=".2f")
```

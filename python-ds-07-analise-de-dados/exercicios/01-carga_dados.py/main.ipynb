```python
# Importa pandas para utilizar os métodos .head(), .info() e .describe()
import pandas as pd

# Ccarregando um dataset interno do Seaborn
df = sns.load_dataset('car_crashes')

# O método .head()
print(df.head()) # Retorna as primeiras 5 linhas do dataframe
```

COLUNA    -      SIGNIFICADO

total:           Número de motoristas envolvidos em acidentes fatais por bilhão de milhas percorridas.

speeding:        Porcentagem de motoristas envolvidos em acidentes fatais que estavam acima do limite de velocidade.

alcohol:         Porcentagem de motoristas envolvidos em acidentes fatais que estavam sob influência de álcool.

not_distacted:   Porcentagem de motoristas envolvidos em acidentes fatais que não estavam distraídos.

not_previous:    Porcentagem de motoristas envolvidos em acidentes fatais que não tinham acidentes anteriores registrados.

ins_premium:     Valor médio do prêmio do seguro de automóvel no estado.

ins_losses:      Perdas incorridas pelas companhias de seguros por motorista segurado no estado.

abbrev:          Abreviação do estado americano (ex: AL, AK, AZ).

- Será que estados onde os motoristas bebem mais (alcohol) têm um índice total de acidentes maior?

- O valor do seguro (ins_premium) é maior em estados com mais acidentes? (Surpreendentemente, a correlação aqui nem sempre é tão direta quanto parece).

- Note que a coluna not_distracted mede quem não estava distraído. Para saber quem estava distraído, você teria que fazer $100 - not\_distracted$.

```python
'''
    5) Agrupamento (Groupby): Agrupe os dados por uma categoria (ex: "Região" ou "Categoria de Produto") e calcule a soma total de uma métrica (ex: "Faturamento").
'''
```

```python
# Importa a biblioteca seaborn e configura o dataset 'titanic'
import seaborn as sns

df = sns.load_dataset('titanic')

df.head() # Mostra as cinco (5) primeiras linhas do dataset
```

```python
# Importação de pandas e utilização do método groupby
import pandas as pd

# Agrupamento por classe, calculando a arrecadação de tarifa
soma_ticket = df.groupby('class')['fare'].sum()

print(soma_ticket)
```

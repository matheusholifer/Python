```python
'''
    2) Cálculos com NumPy: Selecione uma coluna numérica e utilize o NumPy para calcular manualmente a média, mediana e o desvio padrão. Compare com o resultado do .describe() do Pandas.
'''
```

```python
# Importando as ferramentas que serão usadas
import seaborn as sns
import pandas as pd
import numpy as np

# Carregando um dataset interno do Seaborn
df = sns.load_dataset('car_crashes')

# Mostra as 5 primeiras linhas do dataset 'car_crashes'
df.head()
```

```python
coluna_total = df.total # Número de motoristas envolvidos em acidentes fatais por bilhão de milhas percorridas.

# Mostra a coluna selecionada
print(coluna_total)
```

```python
# Soma os valores númericos da coluna com numpy
soma = np.sum(coluna_total)

# Conta a quantidade de itens usando len()
quantidade = len(coluna_total)

# Cálculo da média manualmente
media = soma / quantidade

# Imprime a média com 2 (Duas) casas decimais
print(f"\nA média é = {media:.2f}")
```

```python
# Cálculo da Mediana com numpy
mediana = np.median(coluna_total)

# Imprime a mediana com 2 (Duas) casas decimais
print(f"\nA mediana é = {mediana:.2f}")
```

```python
# Cálculo do Desvio Padrão com numpy
desvio_padrao = np.std(coluna_total)

# Desvio Padrão Amostral
desvio_amost = np.std(coluna_total, ddof=1)

print(f"\nDesvio Padrão = {desvio_padrao}")
print(f"\nDesvio Padrão Amostral = {desvio_amost}")
```
